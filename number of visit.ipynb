```python
import geopandas as gpd

# 读取公园数据 (GeoJSON 或 Shapefile)
parks_file = '/content/drive/MyDrive/filtered_park_df.geojson'  # 替换为你的公园文件路径
parks_gdf = gpd.read_file(parks_file)
#print(parks_gdf.columns)

# 读取 mobility 数据 (Shapefile)
mobility_file = '/content/drive/MyDrive/上海Grid.shp'  # 替换为mobility文件路径
mobility_gdf = gpd.read_file(mobility_file)

# 查看数据结构
#print(parks_gdf)
print(mobility_gdf.head())
```

```python
# 检查两者的 CRS
print("Mobility CRS:", mobility_gdf.crs)
print("parks CRS:", parks_gdf.crs)
```

```python
parks_gdf = parks_gdf.to_crs("EPSG:32650")

# 检查设置后的CRS
print("Updated CRS:", parks_gdf.crs)
```

```python
import chardet

# 检测文件编码
with open('/content/drive/MyDrive/Shanghai_CITY_PARK.geojson', 'rb') as f:
    result = chardet.detect(f.read())
print(result)

# 根据检测结果调整编码
parks_gdf = gpd.read_file('/content/drive/MyDrive/Shanghai_CITY_PARK.geojson', encoding=result['encoding'])
print(parks_gdf)
```

```python
# 创建 500 米缓冲区

parks_gdf['buffer_500m'] = parks_gdf.geometry.buffer(500)

# 创建新的 GeoDataFrame，仅包含缓冲区几何
buffer_gdf = gpd.GeoDataFrame(parks_gdf[['OBJECTID', 'buffer_500m']], geometry='buffer_500m')

# 检查缓冲区
print(buffer_gdf.head())
```

```python
# 使用空间连接匹配缓冲区和网格
joined = gpd.sjoin(mobility_gdf, buffer_gdf, how='inner', predicate='intersects')


# 按公园统计访问点的数量
visit_counts = joined.groupby('OBJECTID').size().reset_index(name='visit_count')

# 将访问次数合并回公园数据
parks_gdf = parks_gdf.merge(visit_counts, on='OBJECTID', how='left')

# 如果没有访问点，填充为 0
parks_gdf['visit_count'] = parks_gdf['visit_count'].fillna(0)

# 查看结果
print(parks_gdf[['OBJECTID', 'visit_count']])
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you've already calculated the 'visit_count' in the parks_gdf
# parks_gdf contains the 'visit_count' column from the previous calculations

# Step 1: Create the histogram for visit count
plt.figure(figsize=(8, 6))
sns.histplot(parks_gdf['visit_count'], kde=True, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Visit Count to Parks')
plt.xlabel('Visit Count')
plt.ylabel('Number of Parks')
plt.grid(True)
plt.show()
```

```python
# Step 2: Plot the visit count on the map
# Assuming 'visit_count' is a column in parks_gdf
fig, ax = plt.subplots(figsize=(10, 8))

# Plot the parks with a color map based on the visit count
parks_gdf.plot(column='visit_count', ax=ax, legend=True,
               legend_kwds={'label': "Visit Count by Park",
                            'orientation': "horizontal"},
               cmap='YlGnBu')  # 'YlGnBu' is a color map; you can choose other color maps too

plt.title('Visit Count Distribution Across Parks')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
