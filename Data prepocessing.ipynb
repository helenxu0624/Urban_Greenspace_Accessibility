```python
import geopandas as gpd
# Load your point and polygon layers as GeoDataFrames
polygons = gpd.read_file('/content/drive/MyDrive/Capstone/data/上海公园.gpkg')

print(polygons.crs)

# # 将 points 图层的 CRS 转换为 EPSG:4326
polygons = polygons.to_crs("EPSG:32651")
```
